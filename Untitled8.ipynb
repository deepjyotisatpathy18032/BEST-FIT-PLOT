```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


def func(r,a,b,c,d):
    return a+b*r**1+c*r**2+d*r**3
r= np.linspace(0,4,5000)



x = np.array([1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019])
y= np.array([16517,1003,600,800,400,3200,200,12754,4100,12100,12210,5100,12561,15535,28292,18860,50222,75808,40571,99913,129166,188401,101192,5504])



popt, pcov = curve_fit(func, x, y)
print(popt)


plt.figure()
plt.plot(x, y, 'ko', label="Original Noised Data")
E=np.linspace(min(x),max(x),50)
plt.plot(E, func(E, *popt), 'r-', label="Fitted Curve")
plt.legend()
plt.show()


```
